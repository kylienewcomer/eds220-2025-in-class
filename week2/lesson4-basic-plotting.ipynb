## Select ros and columns simutaneously

Can be done using `loc` (labels) or 1iloc` (positions)

### ... by labels or conditions

Syntax:
```
df.loc[row-selection, column-selection]
```


### Example
Select winter surveys for Mugu Lagoon and Tijuana Estuary

````python
df.loc[df['year']>2020, ['MUL_winter', "TJE_winter"]]
````

### ... by position
use iloc()

Syntax:
    
``` 
df.iloc[row-indices, column-indices]
```

### Example
Suppose we want to select rows 3-7 (including 7) and forth and fifth columns

````python
df.iloc[ 3:8, [3,4]]
````

````python
# Read in data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
````

````python
pen_bill = penguins[['bill_length_mm', 'bill_depth_mm']]
print(pen_bill)
````

````python
pen_bill.plot(kind = 'box',
             color = 'darkblue')
````

````python
female_gentoo = penguins[(penguins['sex']== 'female') & (penguins['species']== 'Gentoo')]

female_gentoo['flipper_length_mm'].plot(kind='hist',
                                title='Female Gentoo penguin flipper lengths',
                                xlabel='Flipper length (mm)',
                                grid=True,
                                 color = 'pink')
````
